```python
import pandas as pd
dem = pd.read_csv('vdem.csv')
dem.head()
```

```python
gdp = pd.read_csv('gdp.csv', skiprows=3)
gdp.head() 
```

```python
year_cols = [col for col in gdp.columns if col.isdigit()]

# melt for long format 
gdp_long = (
    gdp
    .melt(
        id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"],
        value_vars=year_cols,
        var_name="year",
        value_name="gdp_growth"
    )
    .assign(year=lambda df: df["year"].astype(int))
)

gdp_long.head()
```

```python
# merge two datasets
dem = dem.rename(columns={"country_text_id": "iso_code"})
gdp = gdp_long.rename(columns={"Country Code": "iso_code"})

df = pd.merge(
    dem,
    gdp,
    on=["iso_code","year"],
    how="left",      
    validate="1:1"   
)
df.head()
```

```python
# filter for years after 1970
df = df[df["year"] >= 1970]
df= df.drop(columns=["Country Name", "Indicator Name", "Indicator Code"])
df.head()
```

```python
df_reg = df[df["year"] >= 1995].copy()

cols_keep = [
    "country_name", "iso_code", "year",
    "v2x_libdem",       
    "gdp_growth",        
    "v2xeg_eqdr", # equality 
    "v2x_rule",   # rule of law
    "v2x_polyarchy"# freedom of expression               
]
df_reg = df_reg[cols_keep]

df_reg = df_reg.sort_values(["iso_code", "year"])
```

```python
import pandas as pd
import json
import statsmodels.api as sm

df_reg["libdem"] = df_reg["v2x_libdem"]
df_reg["eq"]     = df_reg["v2xeg_eqdr"]
df_reg["rule"]   = df_reg["v2x_rule"]
df_reg["poly"]   = df_reg["v2x_polyarchy"]


df_reg["libdem_lag"] = df_reg.groupby("iso_code")["libdem"].shift(1)


df_reg["gdp_growth_t"] = df_reg["gdp_growth"]


df_reg["deq"]    = df_reg.groupby("iso_code")["eq"].diff()
df_reg["drule"]  = df_reg.groupby("iso_code")["rule"].diff()
df_reg["dpoly"]  = df_reg.groupby("iso_code")["poly"].diff()

df_reg.head()
```

```python
df_reg_clean = df_reg.dropna(
    subset=["libdem", "libdem_lag",
            "gdp_growth_t", "deq", "drule", "dpoly"]
)


y = df_reg_clean["libdem"]

X = df_reg_clean[["libdem_lag", "gdp_growth_t",
                  "deq", "drule", "dpoly"]]
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
import json

params = model.params

coeffs = {
    "alpha": float(params["const"]),
    "rho": float(params["libdem_lag"]),
    "b_gdp": float(params["gdp_growth_t"]),
    "b_eq": float(params["deq"]),
    "b_rule": float(params["drule"]),
    "b_poly": float(params["dpoly"]) 
}   

!mkdir -p data

with open("data/coeffs.json", "w") as f:
    json.dump(coeffs, f, indent=2)

BASE_YEAR = int(df_reg["year"].max())

baseline = (
    df_reg[df_reg["year"] == BASE_YEAR]
    [["country_name", "iso_code", "v2x_libdem"]]
    .rename(columns={
        "country_name": "name",
        "iso_code": "code",
        "v2x_libdem": f"libdem_{BASE_YEAR}"
    })
)

baseline.to_json("data/baseline.json", orient="records", indent=2)
baseline.head()
```
